```python
columns_to_drop = [
    "neighborhood_overview",
    "host_since",
    "host_response_time",
    "host_response_rate",
    "host_acceptance_rate",
    "host_thumbnail_url",
    "host_neighbourhood",
    "host_total_listings_count",
    "host_verifications",
    "neighbourhood",
    "calendar_updated",
    "instant_bookable"
]

df = df.drop(columns=columns_to_drop)

df.shape
```

```python
df["price"] = (
    df["price"]
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
)

df["price"] = pd.to_numeric(df["price"])
```

```python
df = df.dropna(subset=["price"])

df.shape
```

```python
df["price"].describe()
```

```python
df = df[(df["price"] >= 25) & (df["price"] <= 3000)]

df["price"].describe()
```
